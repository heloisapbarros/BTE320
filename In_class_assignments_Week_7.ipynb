```python
numbers = []

while True:
  num = float(input("Give a number:"))
  if num == 0:
    break
  numbers.append(num)
numbers.sort(reverse=True)

print(numbers)
```

```python
prices = {
    "P-01": 9.95,
    "P-02": 10.50
}

id = input("Enter product ID: ")

if id in prices:
    print(id, prices[id])
else:
    print(id, "not found")
```

In class assignment 4

```python
names = {}

while True:
    name = input("Give a name: ")
    if name == "":
      break

    if name in names:
      names[name] += 1
    else:
      names[name] = 1

print(names)
```
